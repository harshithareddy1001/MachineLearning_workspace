```python
import pandas as pd
import numpy as np

# Read CSV file
df = pd.read_csv('employee.csv')
import matplotlib.pyplot as plt
 
plt.bar(df["experience"], df["salary"],color='red') #barchart
 
plt.xlabel("Experience")
 
plt.ylabel("Salary")
 
plt.title("Experience vs Salary")
 
plt.show()
```

```python
import matplotlib.pyplot as plt
 
arr=df["age"].value_counts()

 
plt.bar(arr.index, arr.values,color='pink') #barchart
 
plt.xlabel("age")
 
plt.ylabel("count")
 
plt.title("employee age vs count")
 
plt.show()
```

```python
import seaborn as sns
sns.barplot(data=df, x="experience", y="salary", hue='department',color ='green')
```

```python
age_counts = df['age'].value_counts().sort_index()

sns.barplot(
    x=age_counts.index,
    y=age_counts.values,
    color="steelblue",
  
)
plt.yticks(range(1,4))
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.title("Employee Count by Age")
plt.show()
```
