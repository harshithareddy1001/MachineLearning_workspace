```python
average_salary = df['salary'].mean()
df['salary'] = df['salary'].fillna(average_salary)
print(df)
```

```python
df['bonus'] = df['salary'] * 0.10
print(df)
```

```python
df=df.drop(columns=['bonus'])
print(df)
```

```python

def apply_hike(row):
    if row['department'] == 'IT':
        return row['salary'] * 1.10
    elif row['department'] == 'HR':
        return row['salary'] * 1.05
    else:
        return row['salary']
df['new_salary'] = df.apply(apply_hike, axis=1)
print(df)
```

```python
df['updated_salary']=df['salary']
df.loc[df['department'] == 'HR', 'updated_salary'] *= 1.05
```

```python
df['department']=df['department'].replace("HR",1)
df['department']=df['department'].replace("IT",2)
df['department']=df['department'].replace("Finance",3)
df['department']=df['department'].replace("Sales",4)
df['department']=df['department'].replace("Operations",5)
df.to_csv("updated_employees.csv", index=False)

print(df)
```

```python

df_emp = pd.read_csv('updated_employees.csv')

df_dept = pd.read_csv('department.csv')

merged_df = pd.merge(df_emp, df_dept, on='dept_id', how='inner')

merged_df = merged_df[['emp_id', 'name', 'age', 'dept_id', 'dept_name']]

merged_df.to_csv('merged_employee_dept.csv', index=False)

print(merged_df.head())
```
