```python
from ultralytics import YOLO
model = YOLO("yolov8n.pt")  
results = model("people.jpg", show=True)
for r in results:
    for box in r.boxes:
        print(
            "Class:", r.names[int(box.cls)],
            "Confidence:", float(box.conf),
            "x, y cordinates", box.xyxy   # x1,y1,x2,y2
        )
        
```

```python
from ultralytics import YOLO
import cv2
# Load model
model = YOLO("yolov8n.pt")
# Run inference
results = model("people.jpg", conf=0.5, iou=0.4)
for r in results:
    # Print detections
    for box in r.boxes:
        print(
            "Class:", r.names[int(box.cls)],
            "Confidence:", float(box.conf),
            "x, y coordinates:", box.xyxy.tolist()
        )
    # Count objects
    num_objects = len(r.boxes)
    print("Total objects detected:", num_objects)
    # Annotated image (with bounding boxes + labels)
    annotated_img = r.plot()   # returns a numpy array with boxes drawn
    # Add text overlay for count
    cv2.putText(
        annotated_img,
        f"Objects detected: {num_objects}",
        (20, 40), cv2.FONT_HERSHEY_SIMPLEX,
        1, (0, 255, 0), 2
    )
    # Display image
    cv2.imshow("Detections", annotated_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
